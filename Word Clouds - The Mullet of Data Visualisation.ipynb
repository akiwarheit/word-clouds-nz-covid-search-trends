# Exploring Word Clouds: A Cautionary Tale of Ambiguity

I'll be exploring search trends in [New Zealand](https://www.stuff.co.nz/life-style/beauty/120155939/the-enduring-appeal-of-the-mullet-haircut) related to coronavirus using word clouds. Word clouds are a novel data visualisation technique used to depict website metadata tags or visualising freeform text.

Word clouds are quite infamous in the data science community ( [link 1](https://getthematic.com/insights/word-clouds-harm-insights/), [link 2](https://towardsdatascience.com/word-clouds-are-lame-263d9cbc49b7) ). After having played with it for a bit, I've also come to the same conclusion - they don't capture frequency as good as a bar graph, ambiguous, and require a lot of data pre processing to get good results.

I'll be using [Andreas Mueller's](https://github.com/amueller) [WordCloud library.](https://amueller.github.io/word_cloud/index.html) The library is a mix of imperative & declarative programming; imperative in the construction of a `WordCloud`, and declarative that you don't have full control of how the observations are placed in the `WordCloud`.

## Dataset

I grabbed my dataset from [Google Data Trends](https://googletrends.github.io/data/). I'll be using the `Coronavirus_DailyTrendsdata` which has search trends per country separated by tabs. I'll be focusing on the `NZ` tab.

````python
# install using !pip install wordcloud

import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

country = 'NZ'

df = pd.read_excel("Coronavirus_DailyTrendsdata.xlsx", sheet_name=country, header=1)
````

````python
df.head()
````

Looking at the data, the search terms are found in every second column.

````python
c = df.columns

# a series of tuples of column names - e.g. (col1, col2), (col3, col4) etc2
czipped = zip(c[0::2], c[1::2])

idx_cols=[]
val_cols=[]

for idx, val in czipped:
    idx_cols.append(idx)
    val_cols.append(val)

# melt the dataframe using the search terms as values
df = df.melt(id_vars=idx_cols, value_vars=val_cols)

search = df['value'].dropna().apply(lambda x: x.strip()) # array of search terms
````

## Generating the Word Cloud

To better understand how the [API](https://amueller.github.io/word_cloud/generated/wordcloud.WordCloud.html) works, we'll be playing around with the following arguments:

```
colormap - string or matplotlib colormap, default=”viridis”
Matplotlib colormap to randomly draw colors from for each word. Ignored if “color_func” is specified.
-------------------------
background_color color value (default=”black”)
Background color for the word cloud image.
-------------------------
collocations - bool, default=True
Whether to include collocations (bigrams) of two words. Ignored if using generate_from_frequencies.
-------------------------
mask - nd-array or None (default=None)
If not None, gives a binary mask on where to draw words. If mask is not None, width and height will be ignored and the shape of mask will be used instead. All white (#FF or #FFFFFF) entries will be considerd “masked out” while other entries will be free to draw on. [This changed in the most recent version!]
```

[Colormaps](https://matplotlib.org/stable/gallery/color/colormap_reference.html) for fancy colours.

````python
def generate_word_cloud(background_color, colormap, text, collocations = False, height = 600, width = 600):
    return WordCloud(height=height, width=width, random_state=1, background_color=background_color, colormap=colormap, collocations=collocations, stopwords = STOPWORDS).generate(text)

def plot_cloud(wordcloud, title = "Word Cloud", figsize=(100, 70)):
    plt.figure(figsize=figsize)
    plt.title(title, fontsize=240)
    plt.imshow(wordcloud) 
    plt.axis("off");
````

First I'll try to generate a word cloud with `collocations=False`.

> Whether to include collocations (bigrams) of two words. Ignored if using generate_from_frequencies.

````python
background_color='black'
colormap='summer'
text=' '.join(search)

wordcloud = generate_word_cloud(background_color, colormap, text)
plot_cloud(wordcloud)
````

Bigrams we're not included in the example above, leaving gaps. We can try reducing the width & height, but for the sake of demonstrating the library, we'll enable collocations

````python
wordcloud = generate_word_cloud(background_color, colormap, text, True)
plot_cloud(wordcloud)
````

The visualisation above shows lesser gaps and is arguably more appealing. You'll also see observations that are similar. (e.g. nz - new zealand, johns hopkins - john hopkins)

The next example plays with the `mask` argument. This accepts an array that determines the boundaries of the cloud.

````python
import numpy as np
from PIL import Image

im = Image.open("nzorange.png")

mask = np.array(im)

wordcloud = WordCloud(random_state=1, collocations=True, stopwords=STOPWORDS, mask=mask, background_color=background_color, colormap=colormap).generate(text)
plot_cloud(wordcloud, "Masked Word Cloud")
````

One thing to note is that this may lead people to think that the observations above have region significance. This is not the intention and highlights the importance of selecting an appropriate mask.

The examples above have been using `generate` and using raw text as data for the word cloud. As mentioned, Word Clouds are ambiguous and this is exacerbated without proper data processing.

To mitigate ambiguity we can use `generate_with_frequencies`. A quick example below:

````python
# !pip install wonderwords
from wonderwords import RandomWord

freq = {}

r = RandomWord()

for x in np.arange(30):
    w = r.word(include_parts_of_speech=["adjectives"])
    freq[w] = np.random.randint(100)


wordcloud = WordCloud(width=1000, height=800, background_color="green", colormap="binary", relative_scaling=.8, prefer_horizontal=1).generate_from_frequencies(freq)
plot_cloud(wordcloud, "How do you feel today?", (100,80))
````

````python
plt.xticks(np.arange(len(list(freq.keys()))), list(freq.keys()), rotation=90)
plt.plot(freq.values())
````
